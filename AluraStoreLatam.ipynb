### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Calcular la suma de la columna "Precio" como ingreso total estimado
ingresos_tienda1 = tienda["Precio"].sum()
ingresos_tienda2 = tienda2["Precio"].sum()
ingresos_tienda3 = tienda3["Precio"].sum()
ingresos_tienda4 = tienda4["Precio"].sum()

# Mostrar el resultado en una Tabla
data_ingresos = {
    "Tiendas": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Ingresos Totales": [ingresos_tienda1, ingresos_tienda2, ingresos_tienda3, ingresos_tienda4]
}
```

```python
# Imprimir en pantalla la Tabla general
dt_ingresos = pd.DataFrame(data_ingresos)
print(dt_ingresos)
```

# 2. Ventas por categoría

```python
# Agrupar por categoria y contar la cantidad de ventas por cada tipo

# Tienda 1
ventas_por_categoria1 = tienda["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria1.columns = ["Categoría del Producto", "Cantidad de Ventas"]
ventas_por_categoria1["tienda"] = "Tienda 1"

# Tienda 2
ventas_por_categoria2 = tienda2["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria2.columns = ["Categoría del Producto", "Cantidad de Ventas"]
ventas_por_categoria2["tienda"] = "Tienda 2"

# Tienda 3
ventas_por_categoria3 = tienda3["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria3.columns = ["Categoría del Producto", "Cantidad de Ventas"]
ventas_por_categoria3["tienda"] = "Tienda 3"

# Tienda 4
ventas_por_categoria4 = tienda4["Categoría del Producto"].value_counts().reset_index()
ventas_por_categoria4.columns = ["Categoría del Producto", "Cantidad de Ventas"]
ventas_por_categoria4["tienda"] = "Tienda 4"
```

```python
# Unir todos los resultados
ventas_por_categoria  = pd.concat([ventas_por_categoria1, ventas_por_categoria2, ventas_por_categoria3, ventas_por_categoria4], ignore_index=True)

# Mostrar los Resultados
print(ventas_por_categoria)
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificacón del promedio por tienda
calificacion_tienda1 = tienda["Calificación"].mean()
calificacion_tienda2 = tienda2["Calificación"].mean()
calificacion_tienda3 = tienda3["Calificación"].mean()
calificacion_tienda4 = tienda4["Calificación"].mean()

# Mostrar Tabla de Resultados
data_calificacion = {
    "Tiendas": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Promedio de Satisfacción": [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
}
```

```python
# Imprimir en pantalla la Tabla
dt_calificacion = pd.DataFrame(data_calificacion).round(2)
print(dt_calificacion)
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# Función para identificar los produxtos más y menos vendidos
def productos_destacados_por_tienda(dt, nombre_tienda):
  conteo = dt["Producto"].value_counts()
  top = conteo.head(1)  # Producto más vendido
  bottom = conteo.tail(1)  # Producto menos vendido

  print(f"\nTienda: {nombre_tienda}")
  print(f"Producto más vendido: {top.index[0]}")
  print(f"Cantidad de ventas: {top.values[0]}")

  # Concatenar Series para graficar
  datos = pd.concat([top, bottom])

  # Graficar
  plt.figure(figsize=(6, 4))
  datos.plot(kind="bar", color=["#40ED0D", "#E52020"])
  plt.title(f"{nombre_tienda} - Productos más y menos vendidos")
  plt.ylabel("Cantidad de ventas")
  plt.xticks(rotation=45)
  plt.tight_layout()
  plt.show()

# Aplicar para cada tienda
productos_destacados_por_tienda(tienda, "Tienda 1")
productos_destacados_por_tienda(tienda2, "Tienda 2")
productos_destacados_por_tienda(tienda3, "Tienda 3")
productos_destacados_por_tienda(tienda4, "Tienda 4")
```
